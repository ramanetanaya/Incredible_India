**INTRODUCTION**

The Day of the week effect refers to the tendency of the stock to give relatively large returns on a particular day of the week as compared to other days of the week. We analyse five stock indices - NIFTY 50, NIFTY NEXT 50, NIFTY MIDCAP 100, NIFTY SMALLCAP 100, NIFTY BANK to check if such an effect exists in either of these indices. We take weekly data of the above indices and perform Linear Regression and Volatility analyses using dummy varibales for different weekdays to find any relation between returns and day of the week.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from arch import arch_model
````

````python
data=pd.read_csv("NiftyData.csv")
data.head()
````

**Adding Dummy Variables**

We add dummy variables for each of the five Weekdays, so as to generate linear regression models for the stock indices and check for day-of-the-week-effect for different weekdays.

````python
day_dummies=pd.get_dummies(data["DAY"])
day_dummies.head(8)
````

````python
data=pd.merge(left=data,right=day_dummies,left_index=True,right_index=True)
data.drop(["DAY"],axis=1,inplace=True)
data.head()
````



> Indented block



**LINEAR REGRESSION ON THE RETURNS**

Now, we perform Linear regression, considering the weekdays' dummy variables as Independent variables and the Returns of a particular index as the dependent variable. This will give us an idea of the dependence of returns on weekdays from the coefficients and their significance in linear regression.

````python
X=data[["Monday","Tuesday","Wednesday","Thursday","Friday"]]
````

**1. NIFTY 50 RETURNS**



````python
Y_NIFTY50=data.loc[:,["Returns_NIFTY50"]]
MDL_NIFTY50=sm.OLS(Y_NIFTY50,X).fit()
print(MDL_NIFTY50.summary())
````

*The* p-value of Tuesday is 0.008, and it has a positive coefficient in the regression, thus it has some positive effect on the returns.

**2. NIFTY NEXT 50 RETURNS**

````python
Y_NEXT50=data.loc[:,["Returns_NEXT50"]]
MDL_NEXT50=sm.OLS(Y_NEXT50,X).fit()
print(MDL_NEXT50.summary())
````

Here also, the p-value of Tuesday is 0.016, indicating statistical significance similar to the previous case.

**3. NIFTY MIDCAP 100 RETURNS**

````python
Y_MIDCAP=data.loc[:,["Returns_MIDCAP"]]
MDL_MIDCAP=sm.OLS(Y_MIDCAP,X).fit()
print(MDL_MIDCAP.summary())
````

Tuesday has a p-value of 0.02 indicating statistical significance, and a positive coefficient indicating positive effect on returns.

**4. NIFTY SMLCAP 100 RETURNS**

````python
Y_SMLCAP=data.loc[:,["Returns_SMLCAP"]]
MDL_SMLCAP=sm.OLS(Y_SMLCAP,X).fit()
print(MDL_SMLCAP.summary())
````

Here, no regression coefficient is significant at the 5% level. p-value of Tuesday is significant at the 10% level.

***5. NIFTY BANK RETURNS***

````python
Y_BANK=data.loc[:,["Returns_Bank"]]
MDL_BANK=sm.OLS(Y_BANK,X).fit()
print(MDL_BANK.summary())
````

Here also, no coefficient is significant at the 5% level. Tuesday is significant at the 10% level.

For NIFTY50, NIFTY NEXT50, NIFTY MIDCAP 100, we found that returns on Tuesday are generally positive, and statistically significant.

**GARCH MODELS FOR VOLATILITY TESTING**

GARCH stands for "Generalized AutoRegressive Conditional Heteroskedascity". GARCH is a statistical modeling technique used to help predict the volatility of returns on financial assets. GARCH is appropriate for time series data where the variance of the error term is serially autocorrelated following an autoregressive moving average process. Here, we perform GARCH modelling to test for volatility of returns of various indices in the stock market. 

**1.NIFTY 50 Returns**

````python
NIFTY50_GARCH = arch_model(Y_NIFTY50, mean = "Constant" , vol = "GARCH", p = 1, q = 1)
NIFTY50_RESULTS = NIFTY50_GARCH.fit()
NIFTY50_RESULTS.summary()
````

The value of both alpha (coefficient of lagged squared residuals) , beta (coefficient of lagged variance) are both significant. Alpha indicates the volatility's spikiness while beta indicates the volatilty's persistence. 

**2. NIFTY NEXT 50 Returns**

````python
NEXT50_GARCH = arch_model(Y_NEXT50, mean = "Constant" , vol = "GARCH", p = 1, q = 1)
NEXT50_RESULTS = NEXT50_GARCH.fit()
NEXT50_RESULTS.summary()
````

**3. NIFTY MIDCAP 100 Returns**

````python
MIDCAP_GARCH = arch_model(Y_MIDCAP, mean = "Constant" , vol = "GARCH", p = 1, q = 1)
MIDCAP_RESULTS = MIDCAP_GARCH.fit()
MIDCAP_RESULTS.summary()
````

**4. NIFTY SMLCAP 100 Returns**

````python
SMLCAP_GARCH = arch_model(Y_SMLCAP, mean = "Constant" , vol = "GARCH", p = 1, q = 1)
SMLCAP_RESULTS = SMLCAP_GARCH.fit()
SMLCAP_RESULTS.summary()
````

**5. NIFTY BANK Returns**

````python
BANK_GARCH = arch_model(Y_BANK, mean = "Constant" , vol = "GARCH", p = 1, q = 1)
BANK_RESULTS = BANK_GARCH.fit()
BANK_RESULTS.summary()
````

In all the cases, the coefficients of the GARCH(1,1) model are significant. When there is a small value of alpha (NIFTY 50, NEXT 50, MIDCAP, BANKS), and a high value of beta, it indicates that there is persistence of volatility in the indices. In case of NIFTY SMALLCAP 100, we can observe that there value of alpha is more than the other 4 indices, which shows that it is more volatile and unstable than the other four indices.

**CONCLUSION**

GARCH models allow us to look for heteroskedascity in the data. We could see in the above models that volatility is persistent and clustering. We can see that the true parameters all fall within the respective confidence intervals.
